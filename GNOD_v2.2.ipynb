## WebScraping Billboard

```python
# storing webite to scrape in URL
url_billboard = 'https://www.billboard.com/charts/hot-100'
```

```python
# based on these to list build a dataframe
df_billbong = pd.DataFrame({'songs' : songs_billboard, 'artists':artists_billboard})
```

### Working with a Playlist

```python
# import Spotipy Wrapper
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
# initalizing spotipy with credentials and storing in variable sp
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="8cf39186280b40dbb110d4500a43d479",
                                                           client_secret="c156608bc14c460684e5fe13ca86c39f"))
```

```python
#to get all tracks/items from the playlist we create a global function to iterate through the urls

def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']
    while results['next']:
        # print(results['next'])
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
# call the function and apply a playlist
first_pl = get_playlist_tracks('spotify', '4rnleEAOdmFAbRcNCgZMpY')
second_pl = get_playlist_tracks('Cigars & Whiskey', '3WaDPZtv41YEOFmtcKnmbm')
third_pl = get_playlist_tracks('Alex Marty', '6FKDzNYZ8IW1pvYVF4zUN2')
fourth_pl = get_playlist_tracks('Oscar Lallier', '6yPiKpy7evrwvZodByKvM9')
```

```python
#concat all playlists to one
mega_playlist = first_pl + second_pl + third_pl + fourth_pl
```

```python
# 25304 songs in one playlist
len(mega_playlist)
```

```python
# path to get the uri
for i in mega_playlist:
    print(i['track']['id'])
```

```python
# path to get the artists names
for i in mega_playlist:
    print(i['track']['artists'][0]['name'])
```

```python
#path to get song title
for i in mega_playlist:
    print(i['track']['album']['name'])
```

```python
# creating a for loop to gather all infomrations from the JSON we need for a DataFrame. Each loop append the values in a list
# the counter will split the playlist into 100 uri pieces to pull faster the audio features from sp.audio_features

artist_name_pl = []
song_name_pl = []
uri_id_pl = []
counter = 0
audio_f = []

for i in mega_playlist:
    if i["is_local"] == False:
        artist_name_pl.append(i['track']['artists'][0]['name'])
        song_name_pl.append(i['track']['album']['name'])
        uri_id_pl.append(i['track']['uri']) 
        counter = counter+1
        if counter == 100:
            counter = 0
            audio_f.append(sp.audio_features(uri_id_pl))
            uri_id_pl = []
audio_f.append(sp.audio_features(uri_id_pl))
```

```python
# for the moment the audio_f JSON is a list in a list we needs to flaten down one level.
flat_audio_features = [f for feature in audio_f for f in feature]
```

```python
# creating a for loop to extract all needed audio features from the flat_audio_features and store them into lists
loudness = []
speechiness = []
acousticness = []
valence = []
tempo = []
duration_ms = []
danceability = []
energy = []
instrumentalness = []
liveness = []
time_signature = []





for i in flat_audio_features:
    loudness.append(i['loudness'])
    speechiness.append(i['speechiness'])
    acousticness.append(i['acousticness'])
    valence.append(i['valence'])
    tempo.append(i['tempo'])
    duration_ms.append(i['duration_ms'])
    danceability.append(i['danceability'])
    energy.append(i['energy'])
    instrumentalness.append(i['instrumentalness'])
    liveness.append(i['liveness'])
    time_signature.append(i['time_signature'])
```

```python
# all created lists from the frist loop and the second loop are converted into a DataFrame
mega_playlist_df = pd.DataFrame({'artist': artist_name_pl, 
                                 'song' : song_name_pl,
                                 'loudness' : loudness, 
                                 'speechiness': speechiness, 
                                 'acousticness' : acousticness, 
                                 'valence' : valence, 
                                 'tempo' : tempo, 
                                 'duration_ms' : duration_ms,
                                 'danceability' : danceability,
                                 'energy' : energy,
                                 'instrumentalness' : instrumentalness,
                                 'liveness' : liveness,
                                 'time_signature' : time_signature
                                })
```

```python
# a dataframe with 25222 rows and 13 columns
mega_playlist_df.shape
```

## Scaling numericals

```python
# selecting numericals from the mega_playlist_df
numericals = mega_playlist_df.select_dtypes(include = np.number) 
```

```python
# scaling the numericals for processing with the StandardScaler
from sklearn.preprocessing import StandardScaler
numericals_scaled = StandardScaler().fit_transform(numericals)
```

```python
# adding the original column names back to the scaled DataFrame
numericals_scaled_df = pd.DataFrame(numericals_scaled, columns=numericals.columns)
numericals_scaled_df.head(3)
```

## Cluster K-Means 

```python
from sklearn.cluster import KMeans
```

```python
#creating a for loop run a model a range of cluster from 2 until 20 and plot them against the inertias
K = range(2,20)
inertias = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234,
                   )
    kmeans.fit(numericals_scaled_df)
    inertias.append(kmeans.inertia_)
    
import matplotlib.pyplot as plt

plt.figure(figsize=(16,8))
plt.plot(K, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title(f'Elbow Method showing the optimal k within {list(K)}')
```

```python

from sklearn.metrics import silhouette_score

K = range(2,20)
silhouettes = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234,
                   )
    kmeans.fit(numericals_scaled_df)
    silhouettes.append(silhouette_score(numericals_scaled_df,
                                     kmeans.predict(numericals_scaled_df)
                                    )
                   )
    
import matplotlib.pyplot as plt

plt.figure(figsize=(16,8))
plt.plot(K, silhouettes, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title(f'Elbow Method showing the optimal k within {list(K)}')
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

fig, ax = plt.subplots(2, 2, figsize=(15,8))
for k in [2, 3, 4, 5]:
    '''
    Create KMeans instance for different number of clusters
    '''
    km = KMeans(n_clusters=k,
                random_state=1234)
    q, mod = divmod(k, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''
    
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(numericals_scaled_df)
```
